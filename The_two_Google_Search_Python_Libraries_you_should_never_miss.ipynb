# **googlesearch**
It is a Python library used to do an effective Google search. It uses requests and BeautifulSoup4 libraries to scrape the data from Google.

## **Perform basic searching**
Now, with the help of this library, let's search ***Katy Perry***. I mean, why not?. To do this, you need to import the library first and then search for the topic.

````
!pip install googlesearch-python
````

````
import googlesearch
search = googlesearch.search('Katy Perry')
print(search)
````

As soon as you execute the following code, you will be prompted by ***Katy Perry's*** search results as a list. Yes, you heard it right, the return type of the search method is a ***list***.

## **Playing with search method**
You can also control the number of search results as well with `num_results` and also get the results in other languages using `lang` parameters. Let's use the same ***Katy Perry's*** example and limit the search results to ***5*** and get the results in ***french***.

````
import googlesearch
search = googlesearch.search('Katy Perry', num_results = 5, lang = 'fr')
print(search)
````

You know the outcome of the above code. The output will not be displayed in French rather the source they point to will be a French website. And this time there will only be 5 search results displayed.

## **Perform basic searching**
Now with the help of this library, let's search ***Taylor Swift***. I mean, why not?. To do this, you need to import the library first and then search for the topic.

````
!pip install git+https://github.com/abenassi/Google-Search-API
````

````
from googleapi import google
search_results = google.search("Taylor Swift")
print(search_results)
````

As soon as you execute the following code, you will be prompted by ***Taylor Swift's*** search results as a list. The return type of search method is a ***list***.

One of the greatest advantages of using googleapi is such that it prompts not only the `source link` but also appends the result with `name`, and `description` as well. According to the developers, since this is API is in the incremental phase, more features are yet to come.


## **Currency Conversion**
The googleapi also can convert the currency as well. I mean, to be frank, you can use this as the currency converter to convert the desired currency as you want. Let's see how?

For this example, let us convert ***5 USD*** into ***Euros*** and see the result:

````
from googleapi import google
euros = google.convert_currency(5.0, "USD", "EUR")
print("5.0 USD = {0} EUR".format(euros))
````

```
5.0 USD = 3.82350692 EUR
```

Now as expected 5 USD would be now converted to Euros, thanks to the convert_currency function as shown below

Similarly, try feeding in different currencies and see the result.
