## Retrieving the key value as a tuple in aÂ list
For this, we can use the `item()` method that returns a view object. This object contains the key-value pairs as a tuple in a list. Let's see a simple example to understand this concept.


```python
car = {
  "brand": "Mercedes Benz",
  "model": "S - Class",
  "year": 2023
}

x = car.items()

print(x)
```

Now the view object is also dynamic, any changes made to the dictionary will reflect in the view object.

```python
car["year"] = 2024
print(x)
```

Going back to the generator the way we can implement it is by saying

```python
list(d.items())
```

## Randomizing the selected element from the specified sequence
Let's use the `random` module. As the name suggests it's pretty self-explanatory. For example.

```python
import random

print(random.random())
```

In this case, we would write the function as

```python
random.choice(list(d.items()))
```

## Printing the value of the returned dictionary value
A dictionary as discussed above comprises a `key-value` pair. So to return the value you need to write.

```python
key, value = random.choice(list(d.items()))
value
```

Now the value returned would be a dictionary value except for the key. You can always print the `key` to see its return and then map it to the value in the dictionary declared.
